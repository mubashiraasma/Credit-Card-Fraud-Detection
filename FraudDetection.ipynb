## Credit Card Fraud Detection

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from imblearn.over_sampling import SMOTE
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier
```

### Data Loading

```python
df = pd.read_csv("creditcard.csv")
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.duplicated()
```

```python
df.describe()
```

```python
class_distribution = df['Class'].value_counts()
class_distribution
```

```python
labels = ["Non-Fraud", "Fraud"]
plt.pie(class_distribution,labels=labels,autopct='%1.2f%%', startangle=30)
plt.title("Class Distribution")
plt.show()
```

```python
corr_matrix = df.corr()

plt.figure(figsize=(20,10))
sns.heatmap(corr_matrix, cmap='coolwarm', annot=True, fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```

```python
X = df.drop('Class',axis=1)
y = df['Class']
```

```python
X.head()
```

### Scaling of numerical Features

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled = pd.DataFrame(X_scaled, columns=X.columns)
```

#### Splitting the dataset using stratified sampling

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42, stratify=y)
```

#### Applying SMOTE

```python
smote = SMOTE(random_state=42)
X_train_smote, y_train_smote = smote.fit_resample(X_train, y_train)
```

### Creating Models

#### Logistic Regression

```python
lr = LogisticRegression()
```

```python
lr_model = lr.fit(X_train_smote, y_train_smote)
y_predict = lr_model.predict(X_test)
print(classification_report(y_predict, y_test))
```

```python
# Confusion matrix
conf_matrix_lr = confusion_matrix(y_test, y_predict)
sns.heatmap(conf_matrix_lr, annot=True, fmt='d', xticklabels=labels, yticklabels=labels)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Logistic Regression - Confusion Matrix')
plt.show()
```

#### Naive Bayes

```python
naive = GaussianNB()
naive_model = naive.fit(X_train_smote,y_train_smote)
y_predict_naive = naive_model.predict(X_test)
print(classification_report(y_predict_naive, y_test))
```

```python
# Confusion matrix
conf_matrix_naive = confusion_matrix(y_test, y_predict_naive)
sns.heatmap(conf_matrix_naive, annot=True, fmt='d', xticklabels=labels, yticklabels=labels)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Naive Bayes - Confusion Matrix')
plt.show()
```

#### Random Forest Classifier

```python
rfc = RandomForestClassifier(class_weight='balanced', random_state=42)
rfc_model = rfc.fit(X_train_smote, y_train_smote)
```

```python
y_predict_rfc = rfc_model.predict(X_test)
print(classification_report(y_predict_rfc, y_test))
```

```python
confusionMatrix = confusion_matrix(y_predict_rfc, y_test)
sns.heatmap(confusionMatrix, annot=True, fmt='d',
            xticklabels=['Non-Fraud', 'Fraud'], 
            yticklabels=['Non-Fraud', 'Fraud'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

```python
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix

# Initialize the model
svm = SVC(class_weight='balanced', random_state=42)

# Train the model
svm_model = svm.fit(X_train_smote, y_train_smote)

# Predict on test data
y_predict_svm = svm_model.predict(X_test)

# Classification report
print("Classification Report for Support Vector Machine (SVM):")
print(classification_report(y_test, y_predict_svm))

# Confusion matrix
conf_matrix_svm = confusion_matrix(y_test, y_predict_svm)
sns.heatmap(conf_matrix_svm, annot=True, fmt='d', xticklabels=labels, yticklabels=labels)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Support Vector Machine - Confusion Matrix')
plt.show()
```
